```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# filling null values of feature with median of that feature
for feature in [feature for feature in dataset.columns if feature not in ['class']]:
  dataset[feature] = dataset[feature].fillna(dataset[feature].median())
```

```python
X=dataset.drop("class",axis=1)
```

```python
y=dataset["class"]
```

```python
X.info()
```

```python
y
```

```python
y=pd.get_dummies(y,drop_first=True)
y
```

```python
# splitting train test data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# importing the keras library and pakage
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
classifier = Sequential()  # model name classifier
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu',input_dim=170)) # hidden layer 1
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu')) # hidde2 layer 1
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid')) # output layer
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy']) # compiling the model
```

```python
# number of hidden layer and neuron in it is hyperparameter
```

```python
classifier.summary()
```

```python
X_train
```

```python
model_history = classifier.fit(X_train,y_train,batch_size=10,epochs=30,validation_split=0.25)
```

```python
model_history.history.keys()
```

```python
model_history.history['accuracy']
```

```python
# list all data in history 
print(model_history.history.keys())
# summarize history for accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc="upper left")
plt.show()
```

```python
# summarize history for
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model accuracy')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc="upper left")
plt.show()
```

```python
import time
import os

def savedModel_path(model_dir="SAVED_MODELS"):
  os.makedirs(model_dir,exist_ok=True)
  fileName = time.strftime("Model_%Y_%m_%d_%H_%M_%S.h5")
  model_path = os.path.join(model_dir,fileName)
  print(f"your model will be saved at the following location\n{model_path}")
  return model_path
```

```python
classifier.save(savedModel_path())
```

```python
# hyperparameter tuning w.r.t. optimizer
from kerastuner.tuners import RandomSearch
```

```python
def build_model(hp):
  model=Sequential()
  model.add(Dense(units=32,kernel_initializer='he_uniform',activation='relu',input_dim=170))
  model.add(Dense(units=32,kernel_initializer='he_uniform',activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  optimizer=hp.Choice("optimizer",values=["adam","sgd","rmsprop","adadelta"])
  model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
  return model

```

```python
tuner = RandomSearch(build_model,objective="val_accuracy",max_trials=5,directory="project1",project_name='aps-fault')
```

```python
tuner.search(X_train,y_train,validation_data=(X_test,y_test),epochs=5)
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
mymodel=tuner.get_best_models(num_models=1)[0]
```

```python
# hypermater tuning with dif dif layer and neuron
from tensorflow.keras import optimizers
def build_model(hp):
  model=Sequential()
  for i in range(hp.Int('num_layers', 2, 20)):
    model.add(Dense(units=hp.Int('units_' + str(i),min_value=32,max_value=512,step=32),activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(
        optimizer=optimizers.Adam(
            hp.Choice('learning_rate',[1e-2,1e-3,1e-4])),
            loss = 'binary_crossentropy',
            metrics = ['accuracy'])
  return model
```

```python
tuner_second = RandomSearch(build_model,objective = 'val_accuracy',max_trials = 5,directory='project7',project_name='aps-fault-sensor')
```

```python
tuner_second.search(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
