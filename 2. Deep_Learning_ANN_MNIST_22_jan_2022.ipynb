```python
# importing libararies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
import seaborn as sns
```

```python
(X_train_full,y_train_full),(X_test,y_test)=mnist.load_data()
```

```python
# seggregating data for validation
# scqale the data between 0 to 1 by dividing it by 255, as it is unsigned data between 0-255
X_valid,X_train=X_train_full[:5000]/255., X_train_full[5000:]/255
y_valid,y_train=y_train_full[:5000], y_train_full[5000:]
```

```python
# scale the test set as well
X_test=X_test/255
```

```python
print(X_valid.shape)
print(X_train.shape)
print(y_valid.shape)
print(y_train.shape)


```

```python
plt.figure(figsize=(22,22))
sns.heatmap(X_train[1000],annot=True,cmap="binary")
```

```python
LAYERS=[tf.keras.layers.Flatten(input_shape=[28,28],name="inputlayer"),
        tf.keras.layers.Dense(300,activation='relu',name="hiddenlayer1"),
        tf.keras.layers.Dense(100,activation='relu',name="hideenlayer2"),
        tf.keras.layers.Dense(10,activation='softmax',name="outputlayer")]
```

```python
model_clf=tf.keras.models.Sequential(LAYERS)
```

```python
model_clf.summary()
```

```python
LOSS_FUNCTION = "sparse_categorical_crossentropy"
OPTIMIZER="ADAM"
METRICS=["accuracy"]
```

```python
model_clf.compile(loss=LOSS_FUNCTION,optimizer=OPTIMIZER,metrics=METRICS)
```

```python
model_clf.layers[1].name
```

```python
hidden1=model_clf.layers[1]
```

```python
weights,biases=hidden1.get_weights()
```

```python
EPOCHS = 5
VALIDATION_SET=(X_valid,y_valid)
history=model_clf.fit(X_train,y_train,epochs=EPOCHS,validation_data=VALIDATION_SET)
```

```python
import time
import os
def saveModel_path(model_dir="content/SAVED_MODELS/"):
  os.makedirs(model_dir,exist_ok=True)
  fileName = time.strftime("Model_%Y_%m_%d_%H_%S_.h5")
  model_path = os.path.join(model_dir, fileName)
  print(f"your model will be saved at the following location\n{model_path}")
  return model_path
```

```python
model_clf.save(saveModel_path())
```
